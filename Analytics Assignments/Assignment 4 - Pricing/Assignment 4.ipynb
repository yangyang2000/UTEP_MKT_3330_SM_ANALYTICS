# You are consulting for an owner of a convenience store

And you want to figure out the optimal price for some items:

1. Koka-Kola Soda (Normal Price: \$3.99, Wholesale Cost: \$2)
2. Bud's Wiser Beer (Normal Price: \$9.99, Wholesale Cost: \$4)
3. ACME milk (Normal Price: \$4.99, Wholesale Cost: \$2)

You design the following price experiment on 4 consecutive similar weeks.

| Week        | Day        | Item on Discount| Price Discount  |
| :------------| :------------|:---------------| ---------------:|
| 1| Monday| soda | 10% |
| 1| Tuesday| beer      |   15% |
| 1| Wednesday| milk     |    0% |
| 1| Thursday| soda     |    20% |
| 1| Friday| beer     |    10% |
| 1| Saturday| milk     |    15% |
| 1| Sunday| soda     |    0% |
| 2| Monday| beer | 20% |
| 2| Tuesday| milk      |   10% |
| 2| Wednesday| soda     |    15% |
| 2| Thursday| beer     |    0% |
| 2| Friday| milk     |    20% |
| 2| Saturday| soda     |    10% |
| 2| Sunday| beer     |    15% |
| 3| Monday| milk | 0% |
| 3| Tuesday| soda      |   20% |
| 3| Wednesday| beer     |    10% |
| 3| Thursday| milk     |    15% |
| 3| Friday| soda     |    0% |
| 3| Saturday| beer     |    20% |
| 3| Sunday| milk     |    10% |
| 4| Monday| soda | 15% |
| 4| Tuesday| beer      |   0% |
| 4| Wednesday| milk     |    20% |
| 4| Thursday| soda     |    10% |
| 4| Friday| beer     |    15% |
| 4| Saturday| milk     |    0% |
| 4| Sunday| soda     |    20% |

### Each discount above is applied for 3 days. For example, the first Monday, Soda is placed on 10% discount. This 10% discount on soda carries over to Tuesday and Wednesday. The first Tuesday, beer is placed on 15% discount, this discount remains for beer on Wednesday and Thursday.

Assume that prices are rounded to the nearest penny in the dataset (don't make this assumption when computing optimal solutions, though you can round the resulting solutions to make the presentation nicer).

0. Read in the dataset.

1. First, create the column of prices for each of the 3 items taking the above information into account.

2. Assume that the demand follows the constant-elastic form. Run a regression for each product assuming that demands are independent, i.e. the demand for soda only depends on the price of soda. What is the price elasticity of each product?

3. Compute the optimal prices for each product. How do these prices compare to the normal price of each item?

4. Now, relax the assumption that prices of products depend on only own prices, i.e. allow for demand to depend on prices of other products. Also assume that Saturday and Sunday demand for each product might be different since the store typcially gets more traffic on those days. Estimate a demand model for each product. For which products is the assumption of independent demand a good assumption? What products seem to be substitutes/complements of soda? Beer? Milk? Take into consideration the statistical significance when interpreting the parameters.

5. Compute the monopolist's optimal prices by maximizing the joint profit of all 3 products. **Limit the numerical price search to be from the marginal costs of each item to 1.5 times the normal prices** to satisfy the assumption that products have constant elasticity *only* within a small price range. Compute optimal prices for the weekend and weekday separately. How do weekend prices compare to weekday prices? Anything surprising? How do profits compare for weekend and weekdays?

6. Compare these profits to the profits you would have gotten using the prices computed in #3 - be careful of which profit function you use when comparing the 2 prices. Make a logical choice to do an apples to apples comparison for the business. Compare both of these pricing schemes to the profits generated by the previous normal prices.

7. Suppose that the 3 products are sold in separate stores but face the same demand curve. Additionally, each store sets their own price for their product. Given the demand functions and marginal costs are the same and the fixed costs are all 0, what are the equilibrium prices? What is each store's profit? What is the joint profit for all stores? How does this compare to the profit found in #5. Do this for both weekend and weekday demands.

8. Compare the monopolist's profits on each of the items on weekdays and weekends to the weekday and weekend profits of the oligopolists. Do some profits generate more profits in an oligopoly situation?

9. What recommendations would you give the the store owner? How would you frame the prices so that the different prices on weekend and weekday does not seem "unfair?" Any words of caution you want to mention about following the quantitative recommendations?
