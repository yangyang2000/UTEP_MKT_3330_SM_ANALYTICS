# Slicing and dicing a dataset

Notice that the index column doesn't start at 0, like a normal dataset, because this small dataset was taken from a larger dataset and the index columns were kept. 

```python
df[:10] # while this still works
```

```python
df.loc[:10,:] # this does not (the row index 0-10 do not exist)
```

We can fix this indexing by resetting the index and dropping the old index.

```python
df.reset_index(inplace = True) # inplace means it saves the change in the variable df
df.head()
```

```python
df.loc[:10, :]
```

```python
df.drop(['index'], axis = 1, inplace = True) # drops the old index which is now a column named 'index'
```

```python
df.reset_index().drop('index', axis = 1, inplace = True) # this would have worked as well
# but place the inplace argument in the last function's inputs
```

```python
df.columns.values
```

How many properties are there? How many have a pool?

```python
len(df) # len returns number of rows
```

```python
df.poolcnt.unique() # let's look at the possible values for variable poolcnt
```

```python
temp = df[~df.poolcnt.isnull()] # select all the cells for which ~df.poolcnt.isnull() is True
# ~df.poolcnt.isnull() returns True if poolcnt is NOT a null value (no data)
# The ~ "tilde" negates the boolean df.poolcnt.isnull()
len(temp), len(df[df.poolcnt.isnull()]) # we didn't need to create a new variable temp to do this
```

How many have a pool and at least 3 bathrooms?

```python
len(df[(df.poolcnt.isnull())&(df.bathroomcnt<=2)]) # pool is not null + and bathroomcnt>=3
```

Show the columns roomcnt and taxvaluedollarcnt for the 10 highest tax value homes in the dataset that has no pools and 4 bathrooms.

```python
df.loc[(df.poolcnt.isnull())&(df.bathroomcnt==4), 
       ['bedroomcnt','taxvaluedollarcnt']].sort_values(
    by='taxvaluedollarcnt', ascending = False)[:10]

# we can break the code into a few lines by hitting enter after a comma in the code
# look at with this does
# First, we used .loc to slice the data for the no pools and 4 bathrooms conditions 
# We also select the appropriate columns to keep in the .loc index
# Next we used .sort_values to sort the data
# We sort by "taxvaluedollarcnt"
# We specify ascending=False, so that we sort from largest to smallest
# finally, with the sorted dataframe, we select the first 10 rows
```

Let's compute the difference from the average assessed value by cohort year.

1. We need to compute the average value by cohort year.
2. We need to merge this back in to the big dataset.
3. We need to compute the difference in a new column.

```python
temp = df3.groupby('yearbuilt')['taxvaluedollarcnt'].mean().reset_index() # what does this syntax say?
temp.head()
```

```python
#1 compute cohort value average
temp = df3.groupby('yearbuilt')['taxvaluedollarcnt'].mean().reset_index() # what does this syntax say?
temp.rename(columns = {'taxvaluedollarcnt': 'cohort_mean_value'}, inplace= True) # this is an example of renaming the column usinga a dictionary 
# Why did we rename the column? 
temp.head()
```

```python
#2 merge back in to df3
df3 = df3.merge(temp, how = 'left', on = 'yearbuilt')
```

```python
df3.head()
```

```python
df3['delta_value'] = df3.taxvaluedollarcnt-df3.cohort_mean_value
df3.head()
```

# Run a regression

For the cohort year 2010, run the regression:
$$
\text{Value}=\beta_0+\beta_1\text{bdrms}+\beta_2\text{sqft}
$$

```python
temp = df[df.yearbuilt==2010]
```

```python
import statsmodels.formula.api as smf # import statsmodels formula api
res = smf.ols(formula='taxvaluedollarcnt ~ bedroomcnt + calculatedfinishedsquarefeet', data=temp).fit()
```

```python
res.summary()
```

```python
res.params
```

```python
res.params['bedroomcnt'] # reference a particular parameter
```

```python
res.rsquared # reference r^2
```

What about factor variables? Let's treat bedroom as as a factor variable

```python
res = smf.ols(formula='taxvaluedollarcnt ~ C(bedroomcnt) + calculatedfinishedsquarefeet', data=temp).fit()
```

```python
res.summary()
```

```python
res.params
```

```python
res.rsquared
```

```python
res.params['C(bedroomcnt)[T.8]'] # reference a level of bedroom (=9 bedroom)t
```
